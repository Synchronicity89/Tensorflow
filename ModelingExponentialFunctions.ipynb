# Modeling Exponential Functions

```python
import tensorflow as tf
import numpy as np
#tf.logging.set_verbosity(tf.logging.ERROR)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

def myfunc1(x):
    return x**3
def myfunc2(x):
    return 0.2 * x**4 + 0.5 * x**3 + 0.7 * x**2 + 3*x + 17 
def myfunc3(x):
    return x**0.5
```

```python
import numpy as np
x = np.arange(-100, 100, .01)
```

```python
myfunc = myfunc2  #Try other functions too
myfuncv = np.vectorize(myfunc)
y = myfuncv(x)
X = x.reshape(-1, 1)
```

```python
#!pip install sklearn
import sklearn.model_selection as sk
from random import random
X_train, X_test, y_train, y_test = sk.train_test_split(X,y,test_size=0.33, random_state = 42)
```

```python
plt.scatter(X_train,y_train)
```

```python
# Import the kera modules
from tensorflow.keras.layers import Input, Dense, RepeatVector
from tensorflow.keras.models import Model
experiment = "powers" # try classic, underfit (lives up to its name) and powers
#if you have a separate GPU for machine learning keep set it to 1 otherwise 0
#if you have no GPU, change GPU to CPU
with tf.device('/GPU:0'):
    # This returns a tensor. Since the input only has one column
    inputs = Input(shape=(1,))

    # a layer instance is callable on a tensor, and returns a tensor
    # To the first layer we are feeding inputs
    if(experiment == "powers"):
        x = Dense(1, activation='relu')(inputs)**2*random()+(inputs)**3*random()+(inputs)**4*random()
    if(experiment == "classic"):
        x = Dense(32, activation='relu')(inputs)
        # To the next layer we are feeding the result of previous call here it is h
        x = Dense(64, activation='relu')(x)
        x = Dense(64, activation='relu')(x)
    if(experiment == "underfit"):
        # = RepeatVector(64)(inputs)
        x = Dense(2, activation='relu')(inputs)
        x = Dense(4, activation='relu')(x)
        x = Dense(2, activation='relu')(x)
       
    # Predictions are the result of the neural network. Notice that the predictions are also having one column.
    predictions = Dense(1)(x)

    # This creates a model that includes
    # the Input layer and three Dense layers
    model = Model(inputs=inputs, outputs=predictions)
    # Here the loss function is mse - Mean Squared Error because it is a regression problem.
    model.compile(optimizer='rmsprop',
                  loss='mse',
                  metrics=['mse'])
    model.fit(X_train, y_train,  epochs=500, batch_size=100)  # starts training
```

```python
y_test = model.predict(X_test)
plt.scatter(X_test, y_test)
```
